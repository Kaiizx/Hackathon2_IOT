```python
sort_data = data.sort_values(by='time',ascending=True)
sort_data
```

Time Series check

```python
plt.figure(figsize=(22,10))
plt.plot(sort_data.time,sort_data.temp)
plt.title("Temperature by time")
plt.xlabel("Date")
plt.ylabel("Temperature")
plt.show()
```

Linear Regression

```python
x = sort_data.drop(["temp"],axis=1)
y = sort_data['temp']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2,shuffle=False)
X_train.shape, y_train.shape
X_train.head(20)
```

```python
X_train = X_train.drop(['time'],axis=1)
X_test = X_test.drop(['time'],axis=1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
model = LinearRegression()
model.fit(X_train, y_train)

train_score = model.score(X_train, y_train)
test_score= model.score(X_test, y_test)

y_pred = model.predict(X_test)
mae = mean_absolute_error(y_pred,y_test)
```

```python
print('Model Train Score is : ' , train_score)
print('Model Test Score is : ' , test_score)
print('MAE : ', mae)
```
