```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns # visualization library
import matplotlib.pyplot as plt # visualization library
import plotly.express as px
# import the regression module**
from pycaret.regression import *



import os
```

##Cleansing Data

```python
cwd = os.getcwd()
data = pd.read_csv(cwd+'/IOT/IOT_Train.csv')
data
```

```python
drop_list = ['mac','station_name','tambon_code','tambon_namt','amphur_code','amphur_namt','province_code','province_namt','latitude','longitude']
data.drop(drop_list, axis=1,inplace = True)
data
```

```python
data['time'] = pd.to_datetime(data['time'])
data
```

```python
data['time']=data['time'].dt.tz_localize(None)
data['time']=pd.to_datetime(data['time'],format='%Y-%m-%D %H:%M:%S')
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data = data[data['temp'].notna()]
data.drop(['pm10','pm2.5'], axis=1,inplace = True)
data
```

```python
sort_data = data.sort_values(by='time',ascending=True)
sort_data
```

```python
plot_data=sort_data.copy()
```

```python
plot_data['MA'] = data['temp'].rolling(36).mean()
```

```python
fig = px.line(plot_data, x="time", y=["temp","MA"], template = 'plotly_dark')
fig.show()
```

```python
sort_data.drop(['time'],axis=1,inplace = True)
sort_data
```

##Model

```python

from sklearn.model_selection import train_test_split
pc_train,pc_test = train_test_split(sort_data,test_size=0.2,shuffle=False)
pc_train.shape , pc_test.shape
```

```python
s = setup(data = pc_train, test_data = pc_test, target = 'temp', session_id = 123)
```

```python
best = compare_models()
```

##Prepare Test Data

```python
test = pd.read_csv(cwd+'/IOT/IOT_Test.csv')
test
```

```python
test.drop(drop_list, axis=1,inplace = True)
test
```

```python
test_id = test['id']
test.drop(['id','time','pm10','pm2.5'], axis=1,inplace = True)
test
```

##Prediction


```python
predictions = predict_model(best, data=test)
predictions['prediction_label']
```

```python
predictions['prediction_label']
```

```python
test_raw = pd.read_csv(cwd+'/IOT/IOT_Test2.csv')
test_raw['id']=test_id
test_raw['temp']=predictions['prediction_label']
test_raw
test_raw.to_csv(cwd+'/IOT/IOT_Test2.csv',index=False)
```
